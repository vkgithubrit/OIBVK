```python
x=df[["Present_Price","Driven_kms","Fuel_Type","Selling_type","Transmission","Owner","Age"]]
y=df['Selling_Price']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print("Size of training data : ", x_train.shape[0])
print("Size of test data : ", x_test.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
ss.fit(x_train)
x_train = ss.transform(x_train)
x_test = ss.transform(x_test)
```

```python
df.head(1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score 
```

```python
lr =  LinearRegression()
lr.fit(x_train,y_train)
pred1 = lr.predict(x_test)
score1=r2_score(y_test, pred1)
score1
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
pred2=rf.predict(x_test)
score2=r2_score(y_test,pred2)
score2
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gb=GradientBoostingRegressor()
gb.fit(x_train,y_train)
pred3=gb.predict(x_test)
score3=r2_score(y_test,pred3)
score3
```

```python
from xgboost import XGBRegressor
```

```python
xg=XGBRegressor()
xg.fit(x_train,y_train)
pred4=xg.predict(x_test)
score4=r2_score(y_test,pred4)
score4
```

```python
from sklearn.svm import SVR
```

```python
svm=SVR()
svm.fit(x_train,y_train)
pred5=svm.predict(x_test)
score5=r2_score(y_test,pred5)
score5
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
knn.fit(x_train,y_train)
pred6=knn.predict(x_test)
score6=r2_score(y_test,pred6)
score6
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
dt.fit(x_train,y_train)
pred7=dt.predict(x_test)
score7=r2_score(y_test,pred7)
score7
```

```python
df_final = pd.DataFrame({'Models':['LR','RF','GBR','XG','svm','knn','DT'],
             "R2_SCORE":[score1,score2,score3,score4,score5,score6,score7]})
```

```python
df_final
```
